# Data visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes=plt.style.available
print(themes)
```

```python
##line plot
x=np.arange(10)
y1=x**2
y2=2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.style.use("seaborn")
plt.plot(x,y1,color="red",label="Apple",linestyle="dashed",marker='*')
##plt.show() if you need two graphs use it
plt.plot(x,y2,color="green",label="Kiwi",marker='o')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over time")
plt.legend()  ## for indicating label on graph
plt.show()
```

```python
plt.plot? ##for seeing documentation of plot function
```

```python
prices=np.array([1,2,3,4])**3
print(prices)
```

```python
plt.plot(prices) ## one d array plot
plt.show()
```

##scatter plots

```python
## Adjust the Size of Any Plot
plt.figure(figsize=(4,4)) ##size adjustment
plt.style.use("seaborn")
plt.scatter(x,y1,color="red",label="Apple",linestyle="dashed",marker='*')
##plt.show() if you need two graphs use it
plt.scatter(x,y2,color="green",label="Kiwi",marker='^')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter plot -Prices of Fruits over time")
plt.legend()  ## for indicating label on graph
plt.show()
plt.show()
```

##Bar-Graph

```python
plt.bar? ##see the documentation
```

```python
## graph in dark background
plt.style.use("dark_background")##use any themes
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",color="green",tick_label=["Gold","Platinum","Silver"]) ###current year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year",color="yellow") ## next year
plt.title("Metal Price Comparision")
plt.xlim(-1.5,5)
plt.ylim(0,40)
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## Histogram

```python
## create data sets using normal distribution
## use numpy to create standard normal distribution 
## then plot histogram
## x=Xsn*sigma + u
Xsn=np.random.randn(100)
##print(Xsn)
sigma=8
u=70
X=np.round(Xsn*sigma + u)
x2=np.round(Xsn*15 + 40)
print(X,x2)
```

```python
plt.style.use("seaborn")
plt.hist(X,alpha=0.8,label="Physics")
plt.hist(x2,alpha=0.9,label="Maths")
plt.ylabel("Prob/FreqCount of student")
plt.xlabel("Marks range")
plt.title("Histogram")
plt.show()
```

# Seaborn Library

```python
import seaborn as sns
#use official documentry of seaborn for detail knowledge
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# bar graph using seaborn
sns.barplot(x='sex' ,y='total_bill',data=tips,estimator=np.std)
```

```python
##count plot
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
#box plot #dotts are outlier
sns.boxplot(x='day' ,y='total_bill' ,data=tips,hue='smoker')##hue is used to categorised..u
```

```python
##violenplot
# split is true for same bar while false for two different bar as above
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Distribution plot

```python
sns.distplot(tips['total_bill'],kde=True,bins=50)##kde=false will disappear line
```

```python
##kdeplot similar to normal distribution
sns.kdeplot(tips['total_bill'])
```

## Joint Plot

```python
#use to compare two variables
#kind='scatter' is default
sns.jointplot(x='total_bill' ,y='tip' ,data=tips ,kind='reg')
```

## Pair Plot

```python
##compare every variable of dataset in pair
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips)
```
