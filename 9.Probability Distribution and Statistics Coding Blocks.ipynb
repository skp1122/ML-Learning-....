# Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate  a Normal Distribution (1-D)
# plot histogram
u = 5
sigma = 0.5# spread of curve
vals = u + sigma*np.random.randn(1000)

#print(vals.shape)
#print(vals)
plt.hist(vals,10)
plt.show()
```

```python
vals= np.round(vals)#rounding off
z = np.unique(vals,return_counts=True)
print(z)
```

```python
#scatter plot of points generated
#uni-variate normal distrubution
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

```python
 # Multi-variate Normal Distribution ,n-d curve
    
    p(x) =(e^(-1/2)*(x-u)^T*z^(-1)*(x-u))/((2*pi)^m/2*|E|^1/2)
    where, |E|= determinant of covariant matrix
            m = no. of dimension
            (x-u)^T = transpose of matrix (x-u)
```

```python
# bi-variate
# sigma yx,xy,xx,yy=0,0,1,1
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])

plt.show()
```
