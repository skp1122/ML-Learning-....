 # Gradient Descent Update for Linear Regression (single feature)

```python
-> theta = init()
-> J(theta) = 1/m * summation{ [( Y(i)actual - Y(i)predicted)]^2 }
-> update theta using Gradient Algorithm , We will find the value of 
    theta that minimizes our loss function
    
    theta = [theta0 , theta1]
    UPDATE->
    theta0 = theta0 - mu*delta of(j(theta)/delta(theta0)) # mu = how big is change
    theta1 = theta1 - mu*delta of (j(theta)/delta(theta1)) # delta - gradient at that point
    Partial derivatives :-
    1st term-
        delta of (j(theta)/delta theta0) = 2/m summation [ [theta0 + theta1*x(i)] - y(i)]
                                     = 2/m summation [Y^(i)-Y(i)]
    2nd term-
        delta of (j(theta)/delta theta1) = 2/m summation [ Y^(i)-Y(i)]*X(i)
        
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Loading data
X = pd.read_csv('./Linear_x_Train.csv')
y = pd.read_csv('./Linear_Y_Train.csv')

# converting into numpy array
X = X.values
y = y.values

# Normalisation
u = X.mean()
std = X.std()

X = (X-u)/std
# visualise the data
plt.scatter(X,y,color='orange')
plt.title("Time vs Performance graph")
plt.xlabel("Hardwork")
plt.ylabel("Perfomance")
plt.show()
```

```python
X.shape , y.shape
```

```python
def hypothesis(x,theta):
    # theta = [theta0, theta1]
    y_ = theta[0] + theta[1]*x
    return y_

def gradient(X,Y,theta):
    m = X.shape[0] # no. of examples in dataset
    grad = np.zeros((2,))
    for i in range(m):
        y = Y[i]
        x = X[i]
        y_ = hypothesis(x,theta)
        grad[0] += (y_ - y)
        grad[1] += (y_ - y)*x
    return grad/m

def error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        total_error += (y_ - Y[i])**2
        
    return total_error/m


def gradientDescent(X,Y):
    max_steps=100
    learning_rate = 0.1
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    
    for i in range(max_steps):
        
        #compute grad and error
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        
        # update theta
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        #store theta for visualisation part
        theta_list.append((theta[0],theta[1]))
        
    return theta,error_list,theta_list
```

```python
theta,error_list,theta_list = gradientDescent(X,y)
theta
# Save the data file
np.save("Theta_List.npy",theta_list)
```

```python
# X - Matrix (m x n) 
# x - vector ( single Example with n feature)
def hypothesis(x,theta): # x is vector
    y_ = 0.0
    n = x.shape[0]
    for i in range(n):
        y_ += (theta[i]*x[i])
    return y_

def error(X,Y,theta):
    e = 0.0
    m = X.shape[0]
    
    for i in range(m):
        y = Y[i]
        y_ = hypothesis(X[i] , theta)
        e += (y - y_)**2
    return e/m

def gradient(X,y,theta):
    m,n = X.shape
    
    grad = np.zeros((n,))
    
    # for all values of j
    for j in range(n):
        # sum over all examples
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            grad[j] += ( y_ - y[i])*X[i][j]
    # out of  both loops
    return grad/m

    
def gradient_descent(X,y,learning_rate=0.1,max_epochs=300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epochs):
        e = error(X,y,theta)
        error_list.append(e)
        
        # Gradient descent
        grad = gradient(X,y,theta)
        for j in range(n):
            theta[j] = theta[j] - learning_rate*grad[j]
    
    return theta,error_list
        
    
```

```python
import time
start = time.time()
theta,error_list = gradient_descent(X,y)
end = time.time()
print("Time taken is ", end-start)
```

# Section : Optimising Code using Vectorization
An Efficient Implementation for Linear Regression using Vectorization
    - Avoid loops in the implementation, except gradient descent main loop
    - Use numpy functions like np.sum() , np.dot() which are quite fast and already optimised

```python
def hypothesis(X,theta):  # X is matrix
    return np.dot(X,theta)

def error(X,y,theta):
    error = 0.0
    m = X.shape[0]
    y_ = hypothesis(X,theta)
    e = np.sum((y-y_)**2)
    
    return e/m

def gradient(X,y,theta):
    y_ = hypothesis(X,theta)
    grad = np.dot(X.T,(y_ - y))
    m = X.shape[0]
    
    return grad/m

def gradient_descent(X,y,learning_rate=0.1,max_iters=300):
    
    n = X.shape[1]
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_iters):
        e = error(X,y,theta)
        error_list.append(e)
        
        # Gradient descent
        grad = gradient(X,y,theta)
        theta = theta -learning_rate*grad
        
    return theta,error_list


```

```python
start = time.time()
theta , error_list = gradient_descent(X,y)
end = time.time()
print("Time taken by vectorised code ", end-start)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X,Y):
    
    Y = np.mat(Y) #converting Y into matrix
    firstPart = np.dot(X.T,X)
    secondPart = np.dot(X.T,Y)
    # np.linalg.pinv == psuedo inverse of matrix
    theta = np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
Steps:-
    -> Read and Normalize the Dataset
    -> Generate W for every query point
    -> No training is involved, directly make prediction using
        colosed form solution we derived! 
        {(X'WX)inv * X'WY } where X' is X_ Transpose
    -> Find the best value of Tau(Bandwidth Parameter)[cross validation]
```

```python
def getW(query_point,X,tau):
    M = X.shape[0]
    W = np.mat(np.eye(M))
    
    for i in range(M):
        xi = X[i]
        x = query_point
        W[i,i] = np.exp(np.dot((Xi-x),(xi-x).T)/(-2*tau*tau))
    return W
```

```python
For Normal Distribution :-
   eq 1= P(x) = {1/(sigma*underroot 2*pi)}*e^{(-1/2)*{((x-u)^2)/sigma^2
   eq 2= P(e(i)) = x is replaced by e(i)         ; e(i)=error
   eq 3= y(i) = 0' * x(i) + e(i); 0' = theta transpose
    Hence, y(i) is also coming from normal distrubution with a mean of (0'*x(i))
   eq 4= P(y(i)/x(i);0) = same only power of e changed by {(-1/2)*{(y(i)-0'*x(i))^2}/sigma^2}
```

```python
Concept of Maximum Livelihood Estimate(Probalistic Approach):-
    -> "theta" that maximize the probablity of generating X,Y.
     -> Data is generated through the line X,Y.
        0* = argmax0 P[D;0]  ;0*=theta optimal ,D = dataset
                                p[D;0] = probablity distribution of D parametrised by 0
        D = {y(1)y(2).....y(m)|x(1)....x(m)}
 eq A   P(y(1)....y(m)|x(1).....x(m);0) = Multiplication of P(y(i)|x(i):0) over i=1 to i =m
                                        = Multiplication of eq 4 over i=1 to i=m
 eq B argmax0 L(0) = eq A ; L(0)= Likelihood of theta
        take log in both side of eq B:-
            log(L(0)) = summation {log(1/root(2*pi))-1/2*((y(i)-0'*x(i))^2)/sigma^2}
                Now for maximizing argmax0 we have to minimize 2nd term of above eqn-
                    ==argmin0(1/2)*summation({(y(i)-0'*x(i))^2}/sigma^2)
                    ==here, sigma is constant, so we have to minimize this term-
                                        ==summation({y(i)-0'*x(i)}^2)  = J(0)  
                                       :- hence we have to minimize J(0)
 Hence , for max the probablity that data is generated through 0, you need to minimize J(0) that is , squared error loss function.                                                      
```
